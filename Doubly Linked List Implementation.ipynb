# Doubly Linked List Implementation
## Problem
Implement a doubly linked list.

## Solution 

```python
class List(object):
    """A doubly linked list."""
    
    class Node(object):
        def __init__(self, value):
            self.value = value
            self.next = None
            self.prev = None
    
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def add_front(self, item):
        new_node = List.Node(item)
        
        if not self.head:
            self.head = new_node
            self.tail = new_node
        
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        
        self.size += 1
    
    def add_rear(self, item):
        new_node = List.Node(item)
        
        if not self.head:
            self.head = new_node
            self.tail = new_node
        
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node
            
        self.size += 1
    
    def remove_front(self):
        if self.size == 0:
            raise IndexError('Attempt to remove from an empty list.')
        
        if self.size == 1:
            self.head = None
            self.tail = None
            
        else:
            self.head = self.head.next
            self.head.prev = None
        
        self.size -= 1
            
    
    def remove_rear(self):
        if self.size == 0:
            raise IndexError('Attempt ot remove from an empty list.')
        
        if self.size == 1:
            self.head = None
            self.tail = None
        
        else:
            self.tail = self.tail.prev
            self.tail.next = None
        
        self.size -= 1
    
    def __len__(self):
        return self.size
    
    def __iter__(self):
        self.curr = self.head
        return self
    
    def __next__(self):
        if self.curr is None:
            raise StopIteration
        val = self.curr.value
        self.curr = self.curr.next
        return val
    
    def __str__(self):
        return '[' + ', '.join([str(x) for x in self]) +']'
```

```python
l = List()
l.add_rear(2)
l.add_rear(5)

print(l)

len(l)

l.add_front(7)
print(l)

l.remove_front()

print(l)

l.remove_rear()
print(l)

l.remove_front()
print(l)

l.add_front(3)
print(l)

l.remove_rear()

print(l)
```
